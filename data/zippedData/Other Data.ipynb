```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies_df = pd.read_csv('rotten_tomatoes_movies.csv.gz')
movies_df
```

```python
top_g = movies_df.value_counts('genres').head(20)
#drama still most popular
#different/more specific genre types
```

```python
fig1, ax1 = plt.subplots(figsize=(15,10))
ax1.barh(top_g.index,top_g, color = 'grey')

ax1.set_title('Top 20 Genres', fontsize=25)

ax1.set_ylabel('Genre', fontsize=15)
ax1.set_xlabel('Frequency', fontsize=15);
```

```python
directors_df = movies_df.value_counts('directors').head(20)
directors_df
```

```python
fig2, ax2 = plt.subplots(figsize=(15,10))
ax2.barh(directors_df.index,directors_df, color = 'grey')

ax2.set_title('Top 20 Genres', fontsize=25)

ax2.set_ylabel('Genre', fontsize=15)
ax2.set_xlabel('Frequency', fontsize=15);
```
